```python
data = df[['Yearly_Income','Cust_Spend_Score']]
```

```python
data
```

```python
ss = StandardScaler()
scaled_data = ss.fit_transform(data)
scaled_data = pd.DataFrame(scaled_data,columns=['Yearly_Income','Cust_Spend_Score'])
scaled_data
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters = 4 , random_state = 100)
model
```

```python
model.fit(scaled_data)
```

```python
model.labels_
```

```python
data['lables'] = model.labels_
```

```python
data
```

```python
model.inertia_
```

Inertia measures how well a dataset was clustered by K-Means. It is calculated by measuring the distance between each data point and its centroid, squaring this distance, and summing these squares across one cluster. A good model is one with low inertia AND a low number of clusters ( K ).

```python
model.cluster_centers_
```

```python
a=model.cluster_centers_
```

```python
xcorr = a [:,0]
ycorr = a [:,1]
sns.scatterplot(a,x=xcorr,y=ycorr,color='red',markers=True)
plt.xlabel('X-coordinate')
plt.ylabel('Y-coordinate')
plt.title('Cluster Centers')
plt.legend()
plt.show()
```

```python
pred_model = model.predict(scaled_data)
```

```python
pred_model
```

```python
err = []
for i in range(1,11):
    model = KMeans(n_clusters = i , random_state = 100)
    model.fit(scaled_data)
    err.append(model.inertia_)
```

```python
err
```

```python
plt.plot(range(1,11),err,marker='*')
plt.xlabel('K_cluster')
plt.ylabel('Inertia')
plt.show()
```

```python
from sklearn.metrics import silhouette_score
```

```python
for i in range(2,11):
    model = KMeans(n_clusters = i , random_state = 100)
    model.fit(scaled_data)
    s_score = round(silhouette_score(scaled_data,model.labels_),2)
    print('silhouette_score for ',i,'cluster is',s_score)
```

We can say that the clusters are well apart from each other as the silhouette score is closer to 1.
